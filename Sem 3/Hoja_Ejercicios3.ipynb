## Ejercicio 3

Quicksort

```python
import random as r

lst = [r.randint(0, 50) for _ in range(10)]


def quicksort(lst, ini, fin):
    if ini < fin:
        piv = partition(lst, ini, fin)
        # (1) Divide
        quicksort(lst, ini, piv - 1)
        quicksort(lst, piv + 1, fin)


def partition(lst, ini, fin):
    pivote = lst[fin]
    i = ini - 1
    for j in range(ini, fin):
        if lst[j] <= pivote:
            i += 1
            lst[i], lst[j] = lst[j], lst[i]
    lst[i + 1], lst[fin] = lst[fin], lst[i + 1]
    return i + 1


print('Before:', lst)
quicksort(lst, 0, len(lst) - 1)
print('After:', lst)


```

### Explicación

Es un algoritmo de "divide y vencerás", ya que se divide en dos partes el el arreglo y se realiza la **conquista** al ordenar el arreglo parcialmente.
