# **Algoritmo Quick-Union Ponderado**

```python
import graphviz as gv
```

```python
def drawUF(a):
    dot = gv.Digraph(comment='Nada')
    n = len(a)
    for i in range(n):
        # dot.node(str(i), str(i))
        dot.node(str(i))
    for i in range(n):
        if a[i] >= 0:
            dot.edge(str(i), str(a[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

```python
def find(s, a):
    i = a
    # Se busca la raÃ­z
    while s[i] >= 0:
        i = s[i]
    return i
```

```python
def union(s, a, b):
    pa = find(s, a) # a es el i de p
    pb = find(s, b) # b es el i de q
    if pa == pb: return # a y b ya estan unidos
    print("Valor p =", s[pa], "q =", s[pb], "y pa =", pa, "pb =", pb) 
    if s[pa] < s[pb]:
        s[pa] += s[pb]
        s[pb] = pa
        print ("Ponderacion de p < ponderacion de q, entonces, nuevas ponderaciones para la raiz ", pa, "p =", s[pa], "q =", pa, "(valor de q se actualiza con el valor de p)")
    elif s[pb] < s[pa]:
        s[pb] += s[pa]
        s[pa] = pb
        print ("Ponderacion de p > ponderacion de q, entonces, nuevas ponderaciones para la raiz ", pb,  "p =", s[pb], "q =", pb, "(valor de q toma el valor del q con el que se une)")
    else:
        s[pa] += s[pb]
        s[pb] = pa
        print ("Igual ponderacion de p y q, entonces, nuevas ponderaciones p =", s[pa], "q =", s[pb], "(valor de q se actualiza con el valor de p)")
```

Creamos un arreglo ponderado inicial con valores -1 para todos los elementos del conjunto.

```python
#Asignamos un valor de -1 a cada uno de los 10 item del arreglo de ids 
s = [-1 for i in range(10)]
print(s)
drawUF(s)
```

```python
#Ponderacion es el valor del item
#Unimos 3 y 4, Tienen la misma ponderacion, entonces, la ponderacion de p sera igual a la suma de la ponderacion de p + la ponderacion de q (-1+-1) = -2.  La ponderacion de q es el valor de p
union(s, 3, 4)
print(s)
drawUF(s)
```

```python
#Unimos 4 y 9, Tienen diferente ponderacion, p(-2) y q(-1).  Cambiamos la ponderacion de la raiz de 4 que es 3.
union(s, 4, 9)
print(s)
drawUF(s)
```
