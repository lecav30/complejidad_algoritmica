```python
#Definimos la funcion find
def find(s, a):
    i = a
    # Búsqueda recursiva del padre
    # while s[i] != i:
    #     i = s[i]
    # return i
    if s[i] == i:
        return i
    else:
        return find(s, s[i])
```

```python
#Definimos la funcion union
def union(s, a, b):
    pa = find(s, a)
    #print(pa)
    pb = find(s, b)
    #print(pb)
    # Se asigna la raíz
    s[pa] = pb
    #print(s)
```

```python
#Visializamos una lista de elementos de 0 a 9
s = [i for i in range(10)]
print(s)

# Representan:
s_chr = [chr(i + 97) for i in range(10)]
print(s_chr)
```

```python
import graphviz as gv

def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(chr(e + 97))
    if p != e:
      graph.edge(chr(e + 97), chr(p + 97))
  return graph
```

```python
drawDS(s)
```

```python
# Formar grupos
# G1 = {a, b, d}

union(s, 0, 1)
union(s, 0, 3)

# G2 = {c, f, i}

union(s, 2, 5)
union(s, 2, 8)

# G3 = {e, g, j}

union(s, 4, 6)
union(s, 4, 9)

# G4 = {h}
```
