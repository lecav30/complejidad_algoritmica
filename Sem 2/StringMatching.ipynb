```python
# Ubicar un caracter dentro de un texto
txt1 = "Hoyesviernes31deMarzo"
p = 'viernes'

n = len(txt1)
lp = len(p)
result = []

for i in range(n-1):
    if p == txt1[i:i+lp]:
        result.append(i)

print(result)
```

```python
# Generalizamos la función StringMatch

def string_match(txt, p):
    n = len(txt)
    lp = len(p)
    result = []
    
    for i in range(n - lp + 1):
        if p == txt[i:i+lp]:
            result.append(i)
    return result
```

```python
string_match("upc exigete innova", "gete")
```
