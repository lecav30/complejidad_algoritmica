### Función enumerate

```python
students = ['Juan', 'Miguel', 'Rodrigo']

en_stud = list(enumerate(students))
print(en_stud)
```

```python
# Utilizando un for para acceder a las posiciones de los items de una lista
for k,v in en_stud:
    print("Key: {} - Value: {}".format(k, v))
```

```python
# Utilizando solo la lista
for i in range(len(students)):
    print("Key: {} - Value: {}".format(i, students[i]))
```

# Algoritmo Backtracking & NReinas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Crear tablero

board = [4, 7, 2, 6, 1, 0, 3, 5]

for row, col in enumerate(board):
    print("Hay una reina en la fila", str(row), " y columna", str(col))
```

```python
#Funcion draw para dibujar un tablero

def draw(tablero):
  n = len(tablero)
  #np.zeros devuelve un array del tamaÃ±o y tipo indicados inicializando sus valores con ceros
  b = np.zeros((n, n, 3), dtype=int)
  #print(b)
  b += [255, 128, 80]

  b[::2, ::2] = [255, 225, 120]
  b[1::2, 1::2] = [255, 225, 120]

  _, ax = plt.subplots()
  ax.imshow(b)
  for fil, col in enumerate(tablero):
    ax.text(col, fil, u"\u265b", fontsize=200/n, va="center", ha="center")
  ax.set(xticks=[], yticks=[])
  '''
  La función set(xticks=[], yticks=[]) toma un objeto de lista como argumento.
  Los elementos de la lista indican las posiciones en la acción correspondiente
  donde se mostrarán las marcas. 
  Este método marcará los puntos de datos en las posiciones dadas con marcas.
  '''
```

```python
draw(board)
```

```python
def validate(board, row, col):
    for row_i in range(row):
        col_i = board[row_i]
        delta = row - row_i # fila anterior
        # Verificar las posiciones no válidas para ubicar a la reina sin conflicto
        if col in (col_i, col_i + delta, col_i - delta):
            return False
    return True
```

```python
board1 = [1, 3, -1, -1]
draw(board1)
```

# La función assert

```python
assert validate(board1, 2, 2) == True
# Da error de excepción porque no puede ser True el poder colocar una reina en posición 2, 2
```

```python
assert validate(board1, 2, 0) == True
```

```python
# Generalizando la función

def nqueens(board, row): # recibe lista del tablero y parte de la fila 0
    n = len(board)
    # Condición de row
    if row == n:
        draw(board)
    else:
        for col in range(n):
            if validate(board, row, col):
                board[row] = col
                # Hacemos la llamada recursiva
                nqueens(board, row + 1)
```

```python
# Estado inicial - tablero vacío
n = 4
board2 = [-1]*n
print(board2)
```

```python
nqueens(board2, 0)
```
