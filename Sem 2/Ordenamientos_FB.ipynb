### Ordenamiento por selección (Selection Sort)

```python
def selection_sort(lista):
    n = len(lista)
    for i in range(n - 1):
        min_pos = i
        for j in range(i + 1, n):
            if lista[j] < lista[min_pos]:
                min_pos = j
        # verificar si hay un nuevo menor
        if min_pos != i: # Hacer la permutación
            lista[min_pos], lista[i] = lista[i], lista[min_pos] # cambio
```

```python
unordered_list = [r.randint(0, 50) for i in range(10)]
print("Lista desordenada:", unordered_list)
selection_sort(unordered_list)
print("Lista ordenada:", unordered_list)
```

### Ordenamiento burbuja (Bubble Sort)

```python
def bubble_sort(lista):
    n = len(lista)
    for i in range(n - 1):
        # n - i - 1: Esto se hace debido a que como se compara en caso el lista[j]
        # sea mayor que su elemento siguiente. Entonces, siempre los elementos
        # mayores serán enviados al final.
        for j in range(n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j] #cambio
```

```python
unordered_list = [r.randint(0, 50) for i in range(10)]
print("Lista desordenada:", unordered_list)
bubble_sort(unordered_list)
print("Lista ordenada:", unordered_list)
```
